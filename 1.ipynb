 Умова: 
 
   Два економісти заповнюють документи з власними прогнозами щодо курсу певних коштовних паперів, після чого їх складають
   у спільну папку. Ймовірність відхилення прогнозованого значення курсу від фактичного понад допустиму похибку для першого
   економіста дорівнює 0,1 ; для другого 0,2. Перший економіст заповнив 40 документів, другий 60. Під час перевірки,
   навмання взятий із папки документ виявився з суттєвою похибкою. Визначити ймовірність того, що його склав перший 
   економіст.

Вирішення:

1. Вірогідна кількість хибних документів у першого економіста: 40 * 0,1 = 4
2. Вірогідна кількість хибних документів у другого економіста: 60 * 0,2 = 12
3. Загальна вірогідна кількість хибних документів: 4 + 12 = 16
4. Вірогідність авторства першого економіста рандомного хибного документа: 4 / 16 = 1/4
